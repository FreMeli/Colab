**Objetivo:** El objetivo de este programa es obtener el promedio de una columna dada en un DataFrame.

##Opción 1

```python
import numpy as np #se importa la librería numpy 
import pandas as pd #se importa la librería pandas

df1 = pd.DataFrame(np.random.randint(1,20,size=(20, 4)), columns=list('1234')) #Creamos un data frame de numeros random enteros 
df1 #imprimimos el data frame
```

```python
col=str(input("¿Qué número de columna quieres promediar? ")) #Le pedimos al usuario el número de la columna que quiere promediar
verificador=int(col) #Creamos una variable y le hacemos un cast para hacerla tipo int
if(verificador>4): #Si el usuario mete un número mayor al número de columnas que hay se imprime en pantalla un mensaje de error
  print("Error: Solo hay 4 columnas")
else: #Si mete un número válido entonces calcula el promedio de la columna dada
  print("El promedio es: ", df1[col]. mean ())
```
