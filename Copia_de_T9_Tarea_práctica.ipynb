##2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION,MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
cols=['ESTACION', 'MAGNITUD', 'ANO', 'MES'] #Son las columnas con las que queremos quedarnos
cols.extend([i for i in df if i.startswith('D')]) #de las columnas con las que me quedé les agregré la de los días
df= df[cols]
df
```

##3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
df = df.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name ='VALORES_CONTAMINANTES')
df
```

To make analysis of data in table easier, we can reshape the data into a more computer-friendly form using Pandas in Python. Pandas.melt() is one of the function to do so.

Pandas.melt() unpivots a DataFrame from wide format to long format.

melt() function is useful to massage a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are unpivoted to the row axis, leaving just two non-identifier columns, variable and value.

##4. Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
df["DIA"]= df.DIA.str.strip('D') #.strip elimina cualquier caracter ya sea inicial o final
#Con la línea de arriba esamos eliminando la D al inicio de la columna DIA 
df['FECHA'] = df.DIA.apply(str) + '/' + df.MES.apply(str) + '/' + df.ANO.apply(str) #Concatenamos año, mes y día para crear la columna FECHA
df['FECHA'] = pd.to_datetime(df['FECHA'], format='%d/%m/%Y', errors='coerce', dayfirst=True, yearfirst=False)
df
```

##7. Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
#creamos la función:
def Fun_Emi(estacion, contaminante, fecha1, fecha2):
    return df[(df.ESTACION == estacion) & (df.MAGNITUD == contaminante) & (df.FECHA >= fecha1) & (df.FECHA <= fecha2)].sort_values('FECHA').VALORES_CONTAMINANTES
#Establecemos que df.FECHA se encuentre entre las dos fechas dadas, las ordenamos y mostramos los valores contaminantes ya ordenados.
```

```python
Fun_Emi(56, 8, dt.datetime.strptime('2019/02/12', '%Y/%m/%d'), dt.datetime.strptime('2019/05/17', '%Y/%m/%d'))
```

##10. Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estación indicada.

```python
#La función deberá buscar dentro de las estaciones y contaminantes los datos ingresados para mostrar su descripción 
def Res(estacion, contaminante):
    return df[(df.ESTACION == estacion) & (df.MAGNITUD == contaminante)].VALORES_CONTAMINANTES.describe()

print('Resumen Descriptivo: \n', Res(16, 8),'\n', sep='')
```

##11.Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todas las estaciones.

```python
 def Emi_Medias_Mens(contaminante, anio):
    return df[(df.ANO == anio) & (df.MAGNITUD == contaminante)].groupby(['ESTACION', 'MES']).VALORES_CONTAMINANTES.mean().unstack('MES')
    #.unstack: para que vaya pivoteando entre los meses (1:12)

Emi_Medias_Mens(8, 2019)
```
